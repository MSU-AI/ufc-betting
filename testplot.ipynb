```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
test4 = pd.read_csv('test_4_history.csv')
```

```python
test4
```

```python
plt.plot(test4.index, test4['bankroll'])
# Add dashed red line from start to end point
plt.plot([test4.index[0], test4.index[-1]], [test4['bankroll'].iloc[0], test4['bankroll'].iloc[-1]], 
         'r--', label='Start to End')
plt.ylabel('Bankroll')
plt.xlabel('Number of Games')
plt.title('Bankroll Over Time')
plt.legend()
plt.show()
```
